```python
import glob, os, re
from collections import OrderedDict

docs, ques = [], []
dir_path = "/content/drive/MyDrive/chatbot-pm/dataset/PHS"
filenames, jsonfiles = {}, {}
for filename in sorted(glob.glob(os.path.join(dir_path, '*.txt'))):
    res = list(filter(lambda x: x.isdigit(), filename))
    num = int(''.join(res))
    filenames[num] = [filename]

for jsonfile in sorted(glob.glob(os.path.join(dir_path, '*.json'))):
    splited = jsonfile.split('_')
    res = list(filter(lambda x: x.isdigit(), splited[1]))
    num = int(''.join(res))
    # print(num)
    if num in jsonfiles:
        jsonfiles[num].append(jsonfile)
    else:
        jsonfiles[num] = [jsonfile]


filenames = OrderedDict(sorted(filenames.items()))
jsonfiles = OrderedDict(sorted(jsonfiles.items()))

for num in jsonfiles:
    ques.extend(jsonfiles[num])
print(filenames)

for num in filenames:
    with open(filenames[num][0], "r", encoding = "utf-8") as f:
        docs.append(f.read())
all_text = "\n".join(docs)
f = open(f"{dir_path}/phs.txt", "w", encoding='utf-8')
f.write(all_text)
f.close()
# with open(f"{dir_path}/dvb.txt", 'r', encoding='utf-8') as f:
#   all_text = f.read()
print(ques)

print(all_text)
```

## count input tokens

```python
# !pip install tiktoken
import tiktoken
#encoding in {cl100k_base, p50k_base or r50k_base}
def num_tokens_from_string(string: str, model_name: str) -> int:
    encoding = tiktoken.encoding_for_model(model_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens

num_tokens_from_string(all_text, "gpt-3.5-turbo")
```

```python
def max_question_token_length(questions, model_name="gpt-3.5-turbo"):
    return max(num_tokens_from_string(q,model_name) for q in questions)
max_question_token_length(questions)
```

```python

OPENAI_API_KEY = ""

import os

os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY

from langchain.chains import RetrievalQA
from langchain.document_loaders import TextLoader
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.chat_models import ChatOpenAI
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.prompts import PromptTemplate
import time

model_name = "gpt-3.5-turbo"
prompt_template = """Sử dụng tài liệu được đưa cho, hãy giúp tôi tìm câu trả lời cho các câu hỏi sau. Nếu bạn không biết câu trả lời, cứ nói bạn không biết
{context}
Câu hỏi: {question}
Câu trả lời bằng tiếng việt: """

max_ques_token = max_question_token_length(questions, model_name)
prompt_token = num_tokens_from_string(prompt_template, model_name)
max_answ_token = 1000 # expected lol
chunk_size = 4096 - max_ques_token - prompt_token - max_answ_token
chunk_size = int(chunk_size/3)
loader = TextLoader(f"{dir_path}/phs.txt")
docsearch = None

documents = loader.load()
text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(model_name=model_name, chunk_size=chunk_size, chunk_overlap=0)
texts = text_splitter.split_documents(documents)

embeddings = OpenAIEmbeddings()
# docsearch = Chroma.from_documents(texts, embeddings)

# qa = RetrievalQA.from_chain_type(llm=OpenAI(), chain_type="stuff", retriever=docsearch.as_retriever())
# print("Số token của một text: ", num_tokens_from_string(str(texts[0]), model_name ))
# print("Số token của câu hỏi: ", num_tokens_from_string(query, model_name)) #25
# print("Số token của prompt: ", num_tokens_from_string(prompt_template, model_name)) #103

PROMPT = PromptTemplate(template=prompt_template, input_variables=["context", "question"])
chain_type_kwargs = {"prompt": PROMPT}

# qa = RetrievalQA.from_chain_type(llm=ChatOpenAI(model_name = model_name), chain_type="stuff", retriever=docsearch.as_retriever(search_type="mmr",search_kwargs={'k': 3}), chain_type_kwargs=chain_type_kwargs, return_source_documents=True)
if docsearch is None:
    docsearch = Chroma.from_documents(texts, embeddings)
    qa = RetrievalQA.from_chain_type(llm=ChatOpenAI(model_name = model_name), chain_type="stuff", retriever=docsearch.as_retriever(search_type="mmr",search_kwargs={'k': 3}), chain_type_kwargs=chain_type_kwargs, return_source_documents=True)

else:
    for doc in texts:
        docsearch.add_documents([doc])
        time.delay(2)

```
